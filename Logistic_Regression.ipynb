```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import math
import warnings
warnings.filterwarnings( "ignore" )

col_names = ['x1','x2','result']
#load dataset
df = pd.read_csv("Student-University.csv", names=col_names)
print(df)
```

```python
X = df.iloc[:,:-1].values
print(X)
```

```python
Y = df.iloc[:,-1:].values
print(Y)
```

```python
# Splitting dataset into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
x1 = X_train[:,0]
```

```python
b0=0.0
b1=0.0
b2=0.0
y = b0 + b1*x1 + b2*x2
print(y)
```

```python
prediction = 1/(1+np.exp(-y))
print(np.round(prediction))
```

```python
alpha = 0.3
epochs = 1000
x=1
for epoch in range(epochs):
    prediction = 1/(1+np.exp(-y))
    # Update b0,b1,b2
    b0 = b0 + alpha * (y - prediction) * prediction * (1-prediction) * x
    b1 = b1 + alpha * (y - prediction) * prediction * (1-prediction) * x1
    b2 = b2 + alpha * (y - prediction) * prediction * (1-prediction) * x2
```

```python
y_final = b0 + b1*x1 + b2*x2
prediction = 1/(1+np.exp(-y_final))
print(np.round(prediction))
```
